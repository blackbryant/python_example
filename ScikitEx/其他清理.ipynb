```python
#!pip install yfinance
import pandas as pd
import yfinance as yf

df_quote = yf.download("1101.TW",start="2020-01-01", end="2022-11-30")
df_quote.tail()
df_quote_new =df_quote.resample("M").mean()
df_quote_new
```

```python
df_monthly_sales = pd.read_csv("./stock_monthly_sales.csv")
df_monthly_sales.head()
```

```python
#利用日期join兩個表格
df_quote_new = df_quote.reset_index()
df_quote_new.Date = df_quote_new.Date
df_quote_new.Date = df_quote_new.Date.map(lambda x:str(x)[:4]+str(x)[5:7])
df_quote_new
df_monthly_sales["年月"] = df_monthly_sales["年月"].astype('str')
df = pd.merge(left=df_monthly_sales, right=df_quote_new,left_on="年月",right_on="Date",how="inner")
df_new = df[["Date","單月營收","Adj Close"]]
#欄位改名
df_new.rename({'單月營收':"sales"},axis=1,inplace=True)
df_new

```
